## Eksik Veri
Amaç:
  1. Eksik veri tespiti ve düzeltme
  2. güncel veriyi kaydet

Adımlar:
  1. veri setini yüklemek
  2. eksik veri tespiti
  3. eksik veri oran hesaplama
  4. eksik veri inceleme
  5. eksik veri silme
  6. ortalama ile doldurma
  7. medyan ile doldurma
  8. mod ile doldurma
  9. kategorik verileri mod doldurma
  10. kategorik verileri sabit değer ile doldurma
  11. güncel veri setini oluştur
  12. kaydet

```python
import pandas as pd
```

```python
# 1. veri seti yükleme
df = pd.read_csv("e_ticaret_veri_seti.csv")
```

```python
# 2. eksik verileri tespit et
df.isnull().sum()
```

```python
# 3. eksik veri oranlarını tespit edelim
eksik_sayilari = df.isnull().sum()
eksik_oranlari = 100*(eksik_sayilari / len(df))
eksik_oranlari.round(1)
```

```python
# 4. eksik veri içeren satıları incele
eksik_veri_olan_satirlar = df[df.isnull().any(axis = 1)]
eksik_veri_olan_satirlar.head()
```

```python
# 5. eksik verileri silelim
df_silinen = df.dropna()
df_silinen.shape
```

```python
# 6. sayısal eksik verileri ortalama, median ve mod ile doldurma
df_ortalama = df.copy()
sayisal_sutunlar = df_ortalama.select_dtypes(include=["float64", "int64"]).columns

for sutun in sayisal_sutunlar:
  df_ortalama[sutun] = df_ortalama[sutun].fillna(df_ortalama[sutun].mean())
df_ortalama[sayisal_sutunlar].isnull().sum()
```

```python
# 9. kategorik eskik verileri mod ile doldurma
df_mod_kategorik = df.copy()
kategorik_sutunlar = df_mod_kategorik.select_dtypes(include=["object"]).columns

for sutun in kategorik_sutunlar:
  df_mod_kategorik[sutun] = df_mod_kategorik[sutun].fillna(df_mod_kategorik[sutun].mode()[0])
df_mod_kategorik[kategorik_sutunlar].isnull().sum()
```

```python
# 10. kategorik eksik verileri sabit bir değer ile doldurma
df_sabit_kategorik = df.copy()

for sutun in kategorik_sutunlar:
  df_sabit_kategorik[sutun] = df_sabit_kategorik[sutun].fillna("bilinmiyor")
df_sabit_kategorik[kategorik_sutunlar].isnull().sum()
```

```python
df_guncel= df.copy()

for sutun in sayisal_sutunlar:
  df_guncel[sutun] = df_guncel[sutun].fillna(df_guncel[sutun].median())

for sutun in kategorik_sutunlar:
  df_guncel[sutun] = df_guncel[sutun].fillna(df_guncel[sutun].mode()[0])
```

```python
df_guncel.isnull().sum()
```

```python
# 12. kaydet
df_guncel.to_csv("e_ticaret_veri_seti_eksik_veriler_duzenlendi.csv")
```

## Aykırı Değerler (Outliers)
Amaç:
  1. veri setindeki aykırı değerleri tespit ve inceleme
  2. aykırı değerleri temizleyip güncel veri setini oluştur

Adımlar:
  1. veri setinin yüklenmesi
  2. sayısal sütunlar belirlenir
  3. sayısal sütunların özet istatistikleri
  4. iqr ile aykırı değer sınırlarını hesapla
  5. aykırı değer içeren kayıtları tespit et
  6. aykırı değerleri sütun bazında incele
  7. aykırı değerleri temizle
  8. temizlenmiş veri setinin kaydedilmesi

```python
import pandas as pd
```

```python
# 1. veri setini yükle
df = pd.read_csv("e_ticaret_veri_seti_tekrarlayan_kayitlar_duzenlendi.csv")
df.drop('Unnamed: 0', axis=1, inplace=True)
df.shape
```

```python
# 2. sayısal sütunların tespit edilmesi
sayisal_sutunlar = df.select_dtypes(include = ["int64", "float64"]).columns.tolist()
sayisal_sutunlar
```

```python
# 3. özet istatistik
df.describe()
```

```python
# 4. iqr ile aykırı değer tespiti
aykiri_sinirlar = {}

for sutun in sayisal_sutunlar:

  q1 = df[sutun].quantile(0.25)
  q3 = df[sutun].quantile(0.75)

  iqr = q3 - q1

  alt_sinir = q1 - 1.5 * iqr
  ust_sinir = q3 + 1.5 * iqr

  aykiri_sinirlar[sutun] = {
      "q1":q1,
      "q3":q3,
      "iqr":iqr,
      "alt_sinir":alt_sinir,
      "ust_sinir":ust_sinir
  }
```

```python
aykiri_sinirlar
```

```python
# 5. aykırı değer içeren kayıtları tespit edelim

aykiri_maskesi = pd.Series(False, index=df.index)

for sutun in sayisal_sutunlar:

  alt_sinir = aykiri_sinirlar[sutun]["alt_sinir"]
  ust_sinir = aykiri_sinirlar[sutun]["ust_sinir"]

  sutun_maskesi = (df[sutun] < alt_sinir) | (df[sutun] > ust_sinir)
  aykiri_maskesi = aykiri_maskesi | sutun_maskesi
```

```python
aykiri_kayitlar = df[aykiri_maskesi]
aykiri_kayitlar.head()
```

```python
# 6. aykırı değer sayılarını sutun bazında inceleyelim
for sutun in sayisal_sutunlar:
  alt_sinir = aykiri_sinirlar[sutun]["alt_sinir"]
  ust_sinir = aykiri_sinirlar[sutun]["ust_sinir"]

  aykiri_sayi = ((df[sutun] < alt_sinir) | (df[sutun] > ust_sinir)).sum()
  print(f"{sutun}: {aykiri_sayi}")
```

```python
# 7. aykırı değer içeren kayıtları veri setinden temizleyelim
df_temiz = df[~aykiri_maskesi].copy()

df_temiz.shape
```

```python
df_temiz.to_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

  4. Bar plot ile karşılaştırmalı görselleştirme yapmak
  5. Histogram ile dağılım analizi
  6. Box plot ile aykırı değer tespiti

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. veri setini yükle
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
df["siparis_tarihi"] = pd.to_datetime(df["siparis_tarihi"])
df["siparis_ayi"] = df["siparis_tarihi"].dt.month
```

```python
aylik_toplam_tutar = df.groupby("siparis_ayi")["toplam_tutar"].sum().reset_index()
aylik_toplam_tutar
```

```python
plt.figure(figsize = (6, 3))
plt.plot(aylik_toplam_tutar["siparis_ayi"], aylik_toplam_tutar["toplam_tutar"], marker = "o")
plt.title("Aylara göre toplam tutar")
plt.xlabel("Ay")
plt.ylabel("Toplam tutar")
plt.xticks(aylik_toplam_tutar["siparis_ayi"])
plt.show()
```

```python
# 3. scatter plot
plt.figure(figsize = (6,3))
plt.scatter(df["birim_fiyat"], df["toplam_tutar"])
plt.title("Birim fiyat ile toplam tutar arasındaki ilişki")
plt.xlabel("Birim fiyat")
plt.ylabel("Toplam tutar")
plt.show()
```

```python
# 4. bar plot
kategori_toplam_tutar = (
    df.groupby("kategori")["toplam_tutar"]
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)
kategori_toplam_tutar
```

```python
plt.figure(figsize = (6,3))
sns.barplot(data = kategori_toplam_tutar, x = "kategori", y = "toplam_tutar")
plt.title("Toplam tutara göre kategori dağılımı")
plt.xlabel("Kategori")
plt.ylabel("Toplam tutar")
plt.xticks(rotation = 45)
plt.show()
```

```python
# 5. histogram
plt.figure(figsize = (6,3))
plt.hist(df["toplam_tutar"], bins = 50)
plt.title("Toplam tutar dağılımı")
plt.xlabel("Toplam tutar")
plt.ylabel("Frekans")
plt.show()
```

```python
# box plot
plt.figure()
sns.boxplot(data = df, x = "kategori", y = "toplam_tutar")
plt.title("Toplam tutar dağılımı")
plt.xlabel("Kategori")
plt.ylabel("Toplam tutar")
plt.xticks(rotation = 45)
plt.show()
```

```python
import plotly.express as px

fig = px.box(df, x="kategori", y="toplam_tutar",
             title="Toplam Tutar Dağılımı (Plotly)",
             labels={"kategori": "Kategori", "toplam_tutar": "Toplam Tutar"})

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Plotly grafiğini HTML olarak kaydet
fig.write_html("toplam_tutar_dagilimi_plotly.html")
print("Grafik 'toplam_tutar_dagilimi_plotly.html' olarak kaydedildi.")
```

## Zaman Bazlı Analiz

Amaç:
  1. tarih bilgisi ile zaman bazlı analiz yapma
  2. siparilerin dönemsel dağılımını ve zaman içindeki temel örüntüleri incelemek

Adımlar:
  1. veri setini yükle
  2. sipariş tarihini datetime tipine dönüştür
  3. tarih sütunundan zaman özelliklerini üretmek
  4. aylara göre sipariş sayısını inceleme
  5. aylara göre toplam tutarı inceleme
  6. aylara göre ortalama sipariş tutarı
  7. haftanın günlerine göre sipariş dağılımı
  8. haftanın günelerine göre toplam tutar
  9. En yoğun sipariş dönemlerini belirle
  10. Zaman bazlı analiz sonuçlarını görselleştir

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. veri setini yükle
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
df.shape
```

```python
# siparis_tarihi datetime tipine çevir
df["siparis_tarihi"] = pd.to_datetime(df["siparis_tarihi"])
df["siparis_tarihi"].dtype
```

```python
# 3. tarih  sutunundan zaman özelliklerini türetelim
df["siparis_yili"] = df["siparis_tarihi"].dt.year
df["siparis_ayi"] = df["siparis_tarihi"].dt.month
df["siparis_gunu"] = df["siparis_tarihi"].dt.day
df["haftanin_gunu"] = df["siparis_tarihi"].dt.day_name()
```

```python
# 4. aylara göre sipariş sayısı inceleme
aylik_siparis_sayisi = (
    df.groupby("siparis_ayi")
    .size()
    .reset_index(name = "siparis_sayisi")
    .sort_values("siparis_ayi")
)
aylik_siparis_sayisi
```

```python
# 5. aylara göre toplam tutar
aylik_toplam_tutar = (
    df.groupby("siparis_ayi")["toplam_tutar"]
    .sum()
    .reset_index(name = "toplam_tutar")
    .sort_values("siparis_ayi")
)
aylik_toplam_tutar
```

```python
# siparis sayisi ve toplam tutar arasındaki ilişkiyi anlamak için line plot çizdir
plt.figure(figsize = (6,3))
plt.plot(aylik_siparis_sayisi["siparis_ayi"], aylik_siparis_sayisi["siparis_sayisi"], marker = "o", label = "Sipariş Sayısı")
plt.plot(aylik_toplam_tutar["siparis_ayi"], aylik_toplam_tutar["toplam_tutar"]/1000, marker = "o", label = "Toplam Tutar")
plt.title("Aylara Göre Sipariş Sayısı ve Toplam Tutar")
plt.xlabel("Ay")
plt.ylabel("Sayı")
plt.legend()
plt.show()
```

```python
# 6. adım aylara göre toplam sipariş tutarı
aylik_ortalama_tutar = (
    df.groupby("siparis_ayi")["toplam_tutar"]
    .mean()
    .reset_index(name = "ortalama_siparis_tutari")
    .sort_values("siparis_ayi")
)
aylik_ortalama_tutar
```

```python
# 7. haftanın günlerine göre sipariş dağılımı
gun_sirasi = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
gun_siparis_sayisi = (
    df.groupby("haftanin_gunu")
    .size()
    .reindex(gun_sirasi)
    .reset_index(name = "siparis_sayisi")
)
gun_siparis_sayisi
```

## Değişkenler Arası İlişkiler

Amaç:

  2. heatmap ile görselleştirme

Adımlar:
  1. veri setini yükle
  2. sayısal sütunları belirle
  3. korelasyon matrisini oluştur
  4. heatmap ile görselleştirme

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# veriyi yükle
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
# sayısal sütunları belirle
sayisal_sutunlar = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
# korelasyon matrisi
korelasyon_matrisi = df[sayisal_sutunlar].corr()
korelasyon_matrisi
```

```python
# heatmap ile görselleştirme
plt.figure(figsize = (8,4))
sns.heatmap(korelasyon_matrisi, annot= True, fmt = ".1f", linewidth = 0.5, cmap = "coolwarm");
```

Amaç:

  2. analiz sonuçlarını tablo, grafik ve kısa iş yorumları ile sunmaya çalışalım

Adımlar:
  1. Hangi kategori hem yüksek gelir üretiyor hem de müşteri memnuniyeti açısından güçlü görünüyor
  2. Hangi ödeme türleri daha yüksek ortalama sipariş tutarı üretiyor ve müşteri puanı açısından nasıl performans gösteriyor?
  3. Hangi şehir-kategori kombinasyonları en güçlü fırsatı sunuyor?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
# Hangi kategori hem yüksek gelir üretiyor hem de müşteri memnuniyeti açısından güçlü görünüyor
kategori_raporu = (
    df.groupby("kategori")
    .agg(
        toplam_gelir = ("toplam_tutar", "sum"),
        ortalama_musteri_puani = ("musteri_puani", "mean")
    )
)
kategori_raporu
```

```python
kategori_raporu = kategori_raporu.sort_values(by = ["toplam_gelir", "ortalama_musteri_puani"], ascending = [False, False])
kategori_raporu
```

```python
# kategoriye göre toplam gelir
plt.figure()
sns.barplot(data = kategori_raporu.reset_index(), x = "kategori", y = "toplam_gelir");
```

```python
# kategoriye göre ortalama müşteri puanı
plt.figure()
sns.barplot(data = kategori_raporu.reset_index(), x = "kategori", y = "ortalama_musteri_puani");
```
